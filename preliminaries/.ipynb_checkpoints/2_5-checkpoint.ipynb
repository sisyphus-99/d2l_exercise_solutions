2. **After running the function for backpropagation, immediately run it again and see what happens. Investigate.**

```python
import torch
x = torch.arange(4.0)
x.requires_grad_(True)
y = 2 * torch.dot(x, x)
y.backward()
y.backward()
```

Error: "Trying to backward through the graph a second time (or directly access saved tensors after they have already been freed). Saved intermediate values of the graph are freed when you call .backward() or autograd.grad(). Specify retain_graph=True if you need to backward through the graph a second time or if you need to access saved tensors after calling backward."

In order to reduce storage consumption, all intermediate variables will be deleted when pytorch calls `backward`. Therefore, when backward is called again, it cannot calculate since there are no intermediate variables. In order to prevent the intermediate variables from being deleted, we need to set the parameter retain_graph=True

```python
x = torch.arange(4.0)
x.requires_grad_(True)
y = 2 * torch.dot(x, x)
y.backward(retain_graph=True)
print(x.grad)
y.backward()
print(x.grad)
```

Since the gradient is cumulative, the gradient of x becomes twice as much after the second call.

3. **In the control flow example where we calculate the derivative of `d` with respect to `a`, what would happen if we changed the variable `a` to a random vector or a matrix? At this point, the result of the calculation `f(a)` is no longer a scalar. What happens to the result? How do we analyze this?**

```python
def f(a):
    b = a * 2
    while b.norm() < 1000:
        b = b * 2
    if b.sum() > 0:
        c = b
    else:
        c = 100 * b
    return c
a = torch.randn(size=(2,3), requires_grad=True)
d = f(a)
d.backward()
```

When d becomes a vector or matrix, we cannot calculate its backpropagation. We can use the gradient parameter to make it a scalar:

```python
d.backward(gradient=torch.ones(d.shape))
a.grad
```

4. **Let $f(x) = \sin(x)$. Plot the graph of $f$ and of its derivative $f'$. Do not exploit the fact that $f'(x) = \cos(x)$ but rather use automatic differentiation to get the result.** 

```python
import numpy as np
from d2l import torch as d2l
x = np.linspace(- np.pi,np.pi,100)
x = torch.tensor(x, requires_grad=True)
y = torch.sin(x)
y.sum().backward()

d2l.plot(x.detach(),(y.detach(),x.grad),legend = (('sin(x)',"grad w.s.t x")))
```

7. **Given the graph and the intermediate derivative results, you have a number of options when computing the gradient. Evaluate the result once starting from $x$ to $f$ and once from $f$ tracing back to $x$. The path from $x$ to $f$ is commonly known as *forward differentiation*, whereas the path from $f$ to $x$ is known as backward differentiation.**

- Forward differentiation:
1. input: x=1
2. a = 1, da/dx=2, c = sin1, dc/dx=cos1, d = 1, d(d)/dx=-1
3. b = 0, db/da=1
4. e = 0, de/db=sin1, de/dc=0
5.  f = d + e = 1
6.   df/dx = (2 * 1 * sin1 + cos1 * 0) * 1 + (-1)*1=2sin1-1

- Backward differentiation:
1. input x=1
2. (forward)a = 1,c = sin1,d = 1
3. b = 0
4. e = 0
5. f = 1
6. (backward)df/de = 1, df/d(d) = 1
7. df/db = 1 * sin1,df/dc = 1* 0 = 0
8. df/da= sin1
9. df/dx = 2sin1 -1

8. **When might you want to use forward, and when backward, differentiation? Hint: consider the amount of intermediate data needed, the ability to parallelize steps, and the size of matrices and vectors involved.**

1. Amount of intermediate data needed: Forward mode requires us to store the derivatives, while reverse mode AD only requires storage of the activations.<font color = red>For forward mode, why can't I only store the derivative at the current point. Such as we first get da/dx=2, then db/dx=db/da * da/dx = 2,now we can delete da/dx ?</font>
2. Ability to parallelize steps: Forward differentiation can be parallelized since the intermediate value of each point and the derivative can be calculated at the same time. In backward mode, it does not compute derivatives simultaneously but requires two separate phases. During a forward phase all intermediate variables are evaluated and their values stored in memory. In a following backward phase we then propagates back the derivatives/adjoints with the help of again the chain rule. 
3. Size of matrices and vectors involved:
<font color = red>Why ?</font>
Given a function f:$R^n \to R^m$ we can differentiate between two regimes based on the dimensions to be processed:
- n << m: The input dimension is smaller than the output. The operations applied expand the dimensionality from a set of few units. → Forward mode is computationally cheaper than reverse mode.
- n >> m: The input dimension is larger than the output. This is the classical Deep Learning setting. We obtain a scalar loss evaluation from a forward pass and propagate back to a fairly high-dimensional input. → Reverse mode is computationally cheaper than forward mode.
